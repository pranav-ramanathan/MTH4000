## Outline of a Script

Python is, however, much more versatile than a simple calculator.  It can be used to perform complex operations on a wide variety of data types.  In order to see how this is done, it will be used to consider an example of a Python script, and break down what each of the components is doing.

Take a look at the script below, can you guess what it does without running it?  Try running it, and observe the printed read-out.

```python
n=7
a=0
b=1

for i in range(n):
    c=a+b
    b=a
    a=c
    print(c)
```

It's now clear that this script prints out the first seven Fibonacci numbers.  Over the next few weeks, we will learn exactly how this code works, but for now, let's use this code to highlight important roles within the code.  The first such role is that of a *variable*.

## Variables

It's extremely rare that a computational task can be completed in a single line, and so it may prove advantageous to be able to store values so that they can be used and manipulated in other calculations at a later point.  This is achieved through the use of variables.

A *variable* is a symbol that is used to store a piece of information.  Our code above contained several instances of variables being defined.  For example, consider the very first line:

```python
# this box contains the code for assigning the value 7 to the variable n
# note that the variable n does not need to have been defined before this point

n=7
```

This snippet of code creates a variable `n`, and assigns it the value 7.  This is done using the assignment operator, `=`.  Note that unlike our basic arithmetic calculations above, no value has actually been returned to us here.  Defining a variable does not give an output.  

Once this code box has been run, we can call the value `n` and we will receive the value 7.

```python
# this box calls the value n, meaning that it asks Python to return the value contained in the variable n
# this code requires n to have been defined above

n
```

If we return to our previous example, we can see that there are several instances of variables being defined, e.g. `a=0`, `b=1` and `c=a+b`.

We can combine variables with the numerical operations introduced in the previous section, and perform calculations in terms of our variable:

```python
n**2+12
```

```python
n/(1+n)
```

If we define a new variable, `m`, we can see that it's possible to perform calculations on more than one variable at the same time:

```python
# define a new variable m, and assign it the value 3
m=3

# ask Python to compute the value 7n-4m
7*n-4*m
```

It's also possible to redefine a variable that has already been used above.  Say for instance, changing the value of `n` from 7 to 8.

```python
# redefine n to be the value 8
n=8

# call the (new) value of n
n
```

Note however, that just because we have redefined the value of `n`, it does not immediately update the read-outs of our previous code boxes.  The box above, for instance, should still show a read-out of 37.  If we try running the box again, we find that it updates to include the new value of `n`, and returns the value 44.

Changing the values of variables can cause the outputs of code boxes to change, **but only after they have been rerun**.

## Functions

Certain computational tasks are more common than others.  Take our Fibonacci code above for instance; we only needed to compute the Fibonacci numbers once, but we need to print off data *seven times*.  It would be unfortunate if we had to manually type the code to do this each time, but thankfully, Python provided the `print` function to do it for us.

A function is a data type that accepts one or more objects as inputs (called *arguments*) performs computations using these objects, and then returns some final object as an output.

We will soon learn how to construct functions ourselves, but for now, it's important that you recognise when they're being used.  Take, for instance, the `print` function.  This function is built into Python, and comes pre-loaded into the code's memory.  Just typing the function on its own however, isn't very helpful.

```python
print
```

Python just tells us that this object is a function named `print`, which we already knew.  In order to use a function we must write its name, followed by some input argument in brackets:

```python
print(17.32)
```

Such code is called a *function call*.  Remember, the function is `print` and its argument is 17.32.

It is also possible to use variables as inputs to functions.  Python simply calls the value of the variable and then returns it to the function as an input.

```python
# redefine the variable n to assign it the value of 7
n=7

# call the value of n, and then print that value
print(n)
```

Note that there is a key difference between asking Python to *print* the value of a variable, and asking it to *return* the value.  Can you spot the difference in outputs between the following two code boxes?

```python
# return the value n
n
```

```python
# print the value n
print(n)
```

The code box that *returned* the value `n` gave an output (labelled Out), whereas the code box that *printed* the value `n` did not.  This is a subtle but important difference.  A value being printed is not the same thing as a value being returned.  You can save a returned value for future computations, but you cannot save a printed value.  We will see more of this in future lectures.

Another key difference is that if a code box calls several variables, only the last call is actually returned, whereas any number of values can be printed.  The following code calls the values `n` and `m`:

```python
# call the value n
# since another value is called below, this value will not be returned
n

# call the value m
# since this value is called last, it will be returned
m
```

Whereas this code prints the values `n` and `m`:

```python
# print the value n
print(n)

# print the value m
print(m)
```

As you can see, both values have been printed out beneath the code box.

The print function accepts most data types, including strings:

```python
print('hello world')
```

Notice that even though text strings are always written with apostrophes on either end, these apostrophes disappear when the string is printed.  This does not happen when the string is returned:

```python
'hello world'
```

Finally, the print function accepts more than one argument at a time.  To add more arguments to a function, we separate them with commas.  For example:

```python
# print the value n, then print the value m
print(n,m)
```

```python
# print the text string 'hello world', then print the string 'how are you?'
print('hello world','how are you?')
```

```python
# print the text string 'spam', then print the value n
print('spam',n)
```

This last example shows that we do not need both arguments to have the same data type in order for this function call to be valid.  This is further demonstrated by the following example:

```python
#compute the distance from Edinburgh to London in kilometres
distance_miles=332
print("The distance from Edinburgh to London is", distance_miles, "miles.")
km_per_mile=1.609344
print("There are", km_per_mile, "kilometres in a mile.")
distance_km=distance_miles*km_per_mile
print("So the distance from Edinburgh to London is", distance_km, "kilometres.")
```

## Keywords

There are certain words that Python considers to be special.  They are so special that they cannot be used as variable names, and Python will force an error if you try to assign a value to one of them.  The reason for this is that Python uses these keywords to execute certain fundamental commands.  Such commands are crucial to the structure of a program, and nearly every script you encounter will use at least a couple of them.

In our Fibonacci example, the keywords `for` and `in` are used.  We can see that these words are written on their own, with spaces on either end.  It takes practice to know which keywords are appropriate for a certain context, and this is exactly what we'll be exploring over the few weeks.  In the meantime, we introduce an important keyword, `import`.

### Modules

We have already seen that Python contains many useful built-in functions and data types, but we may need to use something a bit more specialist.  In this case, we import data from *modules*.  A module is a file that contains code that is readable by Python.

If a given module contains useful values, functions and/or data types, then it would be extremely useful to get direct access to those without having to manually open up the file in a word processor.  This is when we use the `import` keyword.

For example, the Python Standard Library contains a module called `math`, which contains mathematical constants and functions.  If you wish to use the value $\pi$, then you can try and ask Python directly, but you will receive an error.

```python
# this does not work
#pi
```

This is because `pi` is not a built-in value for Python.  It is however, defined in the `math` module.  To access it, we use the `import` keyword.

```python
# this does work
import math
math.pi
```

Note that we must prepend the variable `pi` with the name of the module from which it was taken.  

There are a host of other useful values and functions defined in the `math` module.

## Naming Conventions


As mentioned above, there are certain restrictions on which variable names you can and cannot assign values to in Python.  Even in situations where variable names are permitted, they may not always reflect best practice.  Here we briefly discuss the best ways to guarantee that your choice of names results in clear, readable code.

### Variable names to use

You can use any sequence of letters, digits, and underscores (_) to create a variable name, as long as it does not start with a digit. To keep code readable, it makes sense to have names that give some indication of the intended use. Our Fibonacci example above just used integers, so I stuck with the common choices of `n`, `a`, `b`,`c`.

To keep code readable in this module, we will use [naming conventions](http://www.python.org/dev/peps/pep-0008/#naming-conventions): `lower_case_with_single_underscores` variable names and single lowercase letter variable names.

This naming convention makes code as the following easy to read:

```python
#compute the distance from Edinburgh to London in kilometres
distance_miles=332
km_per_mile=1.609344
distance_km=distance_miles*km_per_mile
distance_km
```

### Floating point numbers

Floating point numbers (floats, for short) are real numbers written in decimal form. Python stores floats and integers in different ways, in particular, floats are stored with fixed precision of about 16 digits.

```python
#square root of two
2.0**0.5
```

Note that a float can be equal to an integer, while still maintaining a different data type:

```python
# this is an integer
type(3)
```

```python
# but this is a float
type(3.0)
```

Python understands scientific notation:

```python
1.0e-4
```

```python
1.23e4
```

The computer representation of floating point numbers does not use the decimal number system, but the binary one. When the binary number gets converted to decimal (and vice versa), rounding of the result to 16 digits can result in small errors in the output, as in the following computation, which should have produced an exact result of $534.302208$.

```python
332*1.609344
```

### Complex Numbers

Python can do calculations with complex numbers. A complex number is expressed with using `j` for $j=\sqrt{-1}$ as the imaginary unit. (Yes, I know you are used to $i^2=-1$, but $j$ is sometimes also used, for example in engineering mathematics textbooks.) Alternatively, you can use the built-in `complex()` function with two arguments $x$ and $y$ to give you $x+i\,y$.

```python
complex(3,2)
```

```python
3+2j
```

```python
(1+2j)*(1+1j)
```

Note that we needed to write `1+1j`, and that `1+j` will produce an error.  This is because writing `j` would make Python treat the quantity as a variable.  As per the previous section, variable names are not permitted to start with digits, and so `1j` is unambiguous. 

Note also that the output of a complex calculation is itself of type complex, even if the result is equal to a real number:

```python
(1+2j)*(1-2j)
```

```python
type(5+0j)
```

## Showcasing some advanced examples

The following examples are supposed to get you interested. They show how easy it is to use Python to help you with mathematical problems

And finally, a more advanced example. Here we want to understand the function 
$$f(x,y)=(1-\frac x2+x^5+y^3) \exp(-x^2-y^2)$$
We will do a three-dimensional plot and a contour plot. (Don't worry if this looks too advanced, I just want to wow you with some pretty pictures.)

```python
# define the function - I haven't told you yet how 'def' works, but don't worry
def f(x,y):
    return(1-x/2+x**5+y**3)*numpy.exp(-x**2-y**2)
```

```python
# create an (x,y) grid
n = 256
x = numpy.linspace(-3,3,n)
y = numpy.linspace(-3,3,n)
X,Y = numpy.meshgrid(x,y)
Z = f(X,Y)
```

Having prepared an (x,y) grid and evaluated the function at all points in the grid, we can now produce the desired plots

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
plt.contourf(X, Y, f(X,Y))
plt.show()
```
